```python
# Import Statements
from rembg import remove
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
import os
%matplotlib inline
```

```python
# Make a directory to store original image
os.makedirs('Original', exist_ok= True)
os.makedirs('Masked', exist_ok= True)
```

```python
# Image URL
img_path = "/content/Input_BackgroundRemoval.jpg"
img = Image.open(img_path)
plt.imshow(img)
plt.show()
```

```python
#Let us get the output path
img_name = img_path.split('/')[-1]
output_path = "/content/Masked/"+img_name
output_path
```

```python
# Let us remove input
with open(output_path, 'wb') as f:
  org_img = open(img_path, 'rb').read()
  #output_img = remove(org_img, alpha_matting= True, alpha_matting_foreground_threshold= 50)
  output_img = remove(org_img)
  f.write(output_img)
```

```python
# Display the saved image
img_nobg = Image.open(output_path)
plt.imshow(img_nobg)
plt.show()
```
